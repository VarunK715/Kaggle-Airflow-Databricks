```python
%sql
CREATE OR REPLACE VIEW practice.gold.v_netflix_catalog AS
SELECT 
    s.title,
    s.type,
    s.director,
    s.country,
    s.release_year,
    r.audience,
    r.description AS rating_meaning,
    s.is_current
FROM practice.gold.dim_shows s
LEFT JOIN practice.gold.dim_rating_lookup r 
    ON s.rating = r.rating_code
WHERE s.is_current = true;
```
