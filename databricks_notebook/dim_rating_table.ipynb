```python
from pyspark.sql.types import StructType, StructField, StringType

# 1. Define the data for your ratings
rating_data = [
    ("PG-13", "Teens", "Parents strongly cautioned. May be unsuitable for children under 13."),
    ("TV-MA", "Adults", "For mature audiences. May not be suitable for ages 17 and under."),
    ("PG", "Children", "Parental guidance suggested."),
    ("TV-14", "Teens", "Parents strongly cautioned. May be unsuitable for children under 14."),
    ("TV-PG", "Children", "Parental guidance suggested."),
    ("TV-Y", "Kids", "Suitable for all children."),
    ("TV-Y7", "Kids", "Designed for children age 7 and above."),
    ("R", "Adults", "Restricted. May be unsuitable for children under 17."),
    ("TV-G", "Kids", "Suitable for all ages."),
    ("G", "Kids", "Suitable for all ages."),
    ("NC-17", "Adults", "No one 17 and under admitted."),
    ("NR", "Unknown", "Not Rated."),
    ("Not Rated", "Unknown", "Not Rated."),
    ("UR", "Unknown", "Unrated content.")
]

# 2. Define the Schema
schema = StructType([
    StructField("rating_code", StringType(), False),
    StructField("audience", StringType(), True),
    StructField("description", StringType(), True)
])

# Check if the table already exists in Unity Catalog
if not spark.catalog.tableExists("practice.gold.dim_rating_lookup"):
    print("Table does not exist. Creating dim_rating_lookup...")
    
    # Your existing creation code
    df_ref = spark.createDataFrame(data=rating_data, schema=schema)
    (df_ref.write
      .format("delta")
      .mode("overwrite") 
      .saveAsTable("practice.gold.dim_rating_lookup"))
else:
    print("Table 'practice.gold.dim_rating_lookup' already exists. Skipping creation.")
```
